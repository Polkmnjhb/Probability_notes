# <font color='red'>#Ch00-「前言與簡介」</font>

## #前言—

　　<font color = 'red'>「機率論(Probability Theory)」</font>是一門研究<font color='red'>「隨機現象(Random Phenomena)」</font>的科學，人們將機率定義為一個介於0到1的數，用這個數字代表發生某個現象的可能性有多少，1代表這個現象一定會發生，0代表這個現象絕對不會發生。<br>
　　而隨機現象隨處可見，如擲出一個硬幣三次，在忽略機率微小的例外後，我們能簡單的說出每一次正面或反面朝上的機率均為0.5，代表「正面朝上」和「反面朝上」兩個現象出現的可能性相同，並且可能性的值約為0.5，至於0.5這個值代表什麼、能用來做什麼，我們將會在之後的課程有詳細的介紹。<br>

## #簡介—

<font color='blue'>Chapter01-機率空間(Probability Space)</font> <br>
　　機率空間是由一個序對<font color='red'>($\Omega$,$\mathcal{A}$,$\mathcal{P}$)</font>組成，其中$\Omega$表示<font color='red'>隨機試驗(Random Experiment)</font>其所有可能的<font color='red'>結果(outcome)</font>之集合，也稱$\Omega$為<font color='red'>樣本空間(Sample Space)</font>，每一種可能性即為一個樣本點；$\mathcal{A}$表示由樣本點所成的所有子集合所成的集合；$\mathcal{P}$表示<font color='red'>機率測度(Probability Measure)</font>，$\mathcal{P}$是一個把$\mathcal{A}$中的元素$\mathcal{a}$送到[0,1]區間的函數(i.e.$\mathcal{P}$$：$$\mathcal{a}$$\rightarrow$$[0,1]$)。<br>
　　以上面的試驗為例，丟三次銅版的機率空間為｛$\Omega=｛正正正,正正反,正反正,正反反,反正正,反正反,反反正,反反反｝$,$\mathcal{A}=｛\phi,[正正正],[正正反],...,\Omega｝$,$\mathcal{P}$｝，其中$\mathcal{P}(\Omega)=1$、$\mathcal{P}(｛正正正｝)=\frac18$，依此類推。<br>
　　我們在討論隨機現象時，必須先構造(Construct)一個機率空間，而後才能討論其現象的性質與機率。初學者需要注意的是，兩個不同隨機現象的機率空間不可混用，也就是說，不可用另一個隨機現象的機率測度去計算一個現象的機率。<br>
　　而在本章節中，也會複習<font color='red'>集合論(Set Theory)</font>的相關內容，假設A和B分別是$\Omega$中某一個事件，為了方便起見，我們會以$A^c$代表不包含A的事件，以$A$$\cap$$B$表示同時包含A和B的事件，以$A$$\cup$$B$表示至少包含A或B的事件。最後我們則會介紹<font color='red'>條件機率(Conditional Probability)</font>與<font color='red'>獨立性(Independence)</font>，並舉簡單例子讓讀者了解此兩者的用處為何。

<font color='blue'>Chapter02-離散隨機變數(Discrete Random Variables)</font> <br>
實驗A：一個不均勻的骰子，隨機擲出每一面的可能性分別為$\frac{1}{21}$、$\frac{2}{21}$、$\frac{3}{21}$、$\frac{4}{21}$、$\frac{5}{21}$、$\frac{6}{21}$。<br>
實驗B：將六顆不同顏色不同大小的球放進筒內，隨機抽出紅、黃、綠、藍、靛、紫球的可能性分別為$\frac{1}{21}$、$\frac{2}{21}$、$\frac{3}{21}$、$\frac{4}{21}$、$\frac{5}{21}$、$\frac{6}{21}$。<br>
　　在以上兩個例子中，雖然他們擁有不同的機率空間(因為事件不同)，但若以相同的代號稱呼(如A=擲出點數為1的事件=抽出紅色球的事件,B=點2=黃球,C=點3=綠球,D=點4=藍球,E=點5=靛球,F=點6=紫球)，則擲兩次點數為2的事件，和抽出兩次黃色的事件發生的機率是相同的。為了方便起見，我們將這些情況完全相同的現象分類在一起，並且稱他們擁有相同的<font color='red'>機率分配函數(Probability Distribution Function)</font>。<br>
　　上述例子中，令$\mathcal{X}$是一個隨機變數，則$\mathcal{X}$有六種可能性，分別為A、B、C、D、E、F，也就是說，$\Omega=\{A,B,C,D,E,F\}$，隨機變數可簡單分為<font color='red'>離散型(Discrete)</font>與<font color='red'>連續型(Continuous)</font>。如果$\mathcal{X}$的定義域為有限或無限可數，稱之為DRV(Discrete Random Variables)；如果$\mathcal{X}$的定義域為$\{X:-\infty\lt X \lt\infty\}$，則稱之為CRV(Continuous Random Variables)。<br>
　　本章節中將介紹數個常用的DRV，計算DRV的分配函數，比較各函數間的關係與性質，後段<font color='red'>機率生成函數(Probability Generating Function)</font>與隨機變數之和的分配做總結。

<font color='blue'>Chapter03-離散隨機變數的期望值</font> <br>
　　高中時曾介紹過<font color='red'>期望值(Expectation)</font>的概念，期望值是「隨機試驗在同樣的機會下重複多次的結果計算出的等同「期望」的平均值」，定義為「試驗中每次可能結果的機率乘以其結果的總和」。以擲一個均勻骰子的試驗為例，擲出點數的期望值為 $ E(X)=1\ast(\frac{1}{6})+2\ast(\frac{1}{6})+3\ast(\frac{1}{6})+4\ast(\frac{1}{6})+5\ast(\frac{1}{6})+6\ast(\frac{1}{6}) = \frac{21}{6}$。<br>
　　有了期望值後，便可定義<font color='red'>變異數(Variance)</font>，變異數即在量測所有資料到平均數(=期望值)的平均距離。<br>
　　後段則著重在計算<font color='red'>動差生成函數(Moment Generating Function)</font>與<font color='red'>相關係數(Correlation coefficient)</font>，並以<font color='red'>切比雪夫不等式(Chebyshev's Inequality)</font>結尾。

<font color='blue'>Chapter05-聯合分配隨機變數(Jointly Distributed Random Variables)</font> <br>
　　在同一個機率空間下，有兩個隨機變數$\mathcal{X}$和$\mathcal{Y}$，我們定義<font color='red'>聯合分配函數(Joint Distribution Function)</font>為$F(x,y)=P(\mathcal{X}\leq x , \mathcal{Y}\leq y) , -\infty\lt x,y \lt\infty$<br>
　　在計算$\mathcal{X}$和$\mathcal{Y}$的<font color='red'>邊際密度函數(Marginal Density Function)</font> $f_x$ 和 $f_y$ 後，可藉由聯合分配函數$f(x,y)$ 判斷$\mathcal{X}$和$\mathcal{Y}$是否獨立。<br>
　　本章節也會介紹隨機變數之和的分配、<font color='red'>條件密度函數(Conditional Density Function)</font>和<font color='red'>多變量分佈(Multivariate Distribution)</font>。
