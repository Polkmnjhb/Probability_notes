## #前言—

　　<font color = 'red'>「機率論(Probability Theory)」</font>是一門研究<font color='red'>「隨機現象(Random Phenomena)」</font>的科學，人們將機率定義為一個介於0到1的數，用這個數字代表發生某個現象的可能性有多少，1代表這個現象一定會發生，0代表這個現象絕對不會發生。<br>
　　而隨機現象隨處可見，如擲出一個硬幣三次，在忽略機率微小的例外後，我們能簡單的說出每一次正面或反面朝上的機率均為0.5，代表「正面朝上」和「反面朝上」兩個現象出現的可能性相同，並且可能性的值約為0.5，至於0.5這個值代表什麼、能用來做什麼，我們將會在之後的課程有詳細的介紹。<br>

## #簡介—

<font color='blue'>Chapter01-機率空間(Probability Space)</font> <br>
　　機率空間是由一個序對<font color='red'>($\Omega$,$\mathcal{A}$,$\mathcal{P}$)</font>組成，其中$\Omega$表示<font color='red'>隨機試驗(Random Experiment)</font>其所有可能的<font color='red'>結果(outcome)</font>之集合，也稱$\Omega$為<font color='red'>樣本空間(Sample Space)</font>，每一種可能性即為一個樣本點；$\mathcal{A}$表示由樣本點所成的所有子集合；$\mathcal{P}$表示<font color='red'>機率測度(Probability Measure)</font>，$\mathcal{P}$是一個把$\mathcal{A}$送到[0,1]區間的函數        (i.e.$\mathcal{P}$$：$$\mathcal{A}$$\rightarrow$$[0,1]$)。以上面的試驗為例，丟三次銅版的機率空間為｛$\Omega=｛正正正,正正反,正反正,正反反,反正正,反正反,反反正,反反反｝$,$\mathcal{A}=｛\phi,[正正正],[正正反],...,\Omega｝$,$\mathcal{P}$｝，其中$\mathcal{P}(\Omega)=1$、$\mathcal{P}(｛正正正｝)=\frac18$，依此類推。<br>
　　$\mathcal{A}$
　　我們在討論隨機現象時，必須先構造(Construct)一個機率空間，而後才能討論其現象的性質與機率。初學者需要注意的是，兩個不同隨機現象的機率空間不可混用，也就是說，不可用第二個現象的機率空間去計算第一個現象的機率，反之亦然。<br>
　　而在本章節中，也會複習<font color='red'>集合論(Set Theory)</font>的相關內容，假設A和B分別是$\Omega$中某一個事件，為了方便起見，我們會以$A^c$代表不包含A的事件，以$A$$\cap$$B$表示同時包含A和B的事件，以$A$$\cup$$B$表示至少包含A或B的事件。最後我們則會介紹<font color='red'>條件機率(Conditional Probability)</font>與<font color='red'>獨立性(Independence)</font>，並舉簡單例子讓讀者了解此兩者的用處為何。

<font color='blue'>Chapter02-離散隨機變數(Discrete Random Variables)</font> <br>
實驗A：一個不均勻的骰子，隨機擲出每一面的可能性分別為$\frac{1}{21}$、$\frac{2}{21}$、$\frac{3}{21}$、$\frac{4}{21}$、$\frac{5}{21}$、$\frac{6}{21}$。<br>
實驗B：將六顆不同顏色不同大小的球放進筒內，隨機抽出紅、黃、綠、藍、靛、紫球的可能性分別為$\frac{1}{21}$、$\frac{2}{21}$、$\frac{3}{21}$、$\frac{4}{21}$、$\frac{5}{21}$、$\frac{6}{21}$。<br>
　　在以上兩個例子中，雖然他們擁有不同的機率空間(因為事件不同)，但若以相同的代號稱呼(如A=擲出點數為1的事件=抽出紅色球的事件,B=點2=黃球,C=點3=綠球,D=點4=藍球,E=點5=靛球,F=點6=紫球)，則擲兩次點數為2的事件，和抽出兩次黃色的事件發生的機率是相同的。為了方便起見，我們將這些情況完全相同的現象分類在一起，並且稱他們擁有相同的<font color='red'>機率分配函數(Probability Distribution Function)</font>。<br>
　　上述例子中，令X是一個隨機變數，則X有六種可能性，分別為A、B、C、D、E、F，也就是說，$\Omega=\{A,B,C,D,E,F\}$，隨機變數可簡單分為<font color='red'>離散型(Discrete)</font>與<font color='red'>連續型(Continuous)</font>。
